```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd

# Part 1 - Data pre-processing
# Importing the training set
dataset = pd.read_csv('data.txt')

cleanData = dataset[['Date','HrMn','Spd','Temp','Dewpt','Slp','RHx']]
cleanData = cleanData.iloc[:,:].values

```

```python
dataset
```

```python
cleanData
```

```python
# Re-organizing data for 6 am and 6pm form and cleaning data
trainData = []
lastDate = 0
lastData = []
for data in cleanData:
    if(data[:][1] == 600 and data[:][2] < 200 and data[:][3] < 200 and data[:][4] < 200
       and data[:][5] < 2000 and data[:][6] < 200):
        lastDate = data[:][0] 
        lastData = data
        
    elif(data[:][0] == lastDate and data[:][1] == 1800 and data[:][2] < 200 and data[:][3] < 200 
         and data[:][4] < 200 and data[:][5] < 2000 and data[:][6] < 200):
        if( lastData != []):
            lastData = np.concatenate((lastData, data), axis=0)
            trainData.append(lastData)
            
        lastData = []
        
traing_set = np.asarray(trainData)
```

```python
traing_set
```

```python
# Spliting data into input and output set
X = traing_set[:, 2:7] 
y = traing_set[:,9:14]
```

```python
# Feature Scaling
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X = sc.fit_transform(X)
y = sc.transform(y)
```

```python
# Splitting the dataset into the Training set and Test set
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, random_state = 0)
```

```python
# Reshaping for Keras
X_train = np.reshape(X_train, (803, 1, 5))
X_test = np.reshape(X_test, (201, 1, 5))
```

```python
# Part 2 - Building the RNN

# Importing the Keras libraries and packages
from keras.models import Sequential
from keras.layers import Dense
from keras.layers import LSTM

# Initialising the RNN
regressor = Sequential()

# Adding the input layer and the LSTM layer
regressor.add(LSTM(units = 10, activation = 'sigmoid', input_shape = (None, 5)))

# Adding the output layer
regressor.add(Dense(units = 5))

# Compiling the RNN
regressor.compile(optimizer = 'adam', loss = 'mean_squared_error',metrics=['mae', 'acc'])
```

```python
# Fitting the RNN to the Training set
regressor.fit(X_train, y_train, batch_size = 32, epochs = 100)

```

```python
regressor.evaluate(x= X_test, y= y_test)
```

```python
# Get the predicted result
y_predicted = regressor.predict(X_test)
#y_predicted = sc.inverse_transform(y_predicted)
```

```python
# Part 4 - Evaluating the RNN
import math
from sklearn.metrics import mean_squared_error
rmse = math.sqrt(mean_squared_error(y_test, y_predicted))
print(rmse)
```
